```python
import control as ctrl
import matplotlib.pyplot as plt
```

```python
# Problem 1
# M = 1, B1 = 1, B2 = 1, K = 2
num1 = [1]
den1 = [1, 2, 2]   # s^2 + 2s + 2

system1 = ctrl.TransferFunction(num1, den1)

print("Problem 1 Transfer Function:")
print(system1)
```

```python
poles1 = ctrl.poles(system1)
zeros1 = ctrl.zeros(system1)

print("Problem 1 Poles:", poles1)
print("Problem 1 Zeros:", zeros1)
```

```python
plt.figure()
ctrl.pzmap(system1, grid=True)
plt.title("Pole-Zero Map  Problem 1")
plt.show()
```

```python
# Problem 2
# M = 1, B1 = 1, B2 = 1, K = 1
num2 = [1]
den2 = [1, 2, 1]   # s^2 + 2s + 1

system2 = ctrl.TransferFunction(num2, den2)

print("\nProblem 2 Transfer Function:")
print(system2)
```
