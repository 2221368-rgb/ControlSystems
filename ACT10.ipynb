```python
from openai import OpenAI

# Connect to LM Studio local server
client = OpenAI(
    base_url="http://127.0.0.1:1234/v1",
    api_key="lm-studio"
)

# Initialize the conversation history (This is the AI's memory!)
chat_history = [
    {"role": "system", "content": "You are a helpful, conversational AI assistant."}
]

print("Successfully connected to Gemma and initialized memory!")
print("Chat started! Type 'quit' or 'exit' to stop.\n")

while True:
    # 1. Get your input
    query = input("You: ")
    
    # Check if you want to quit
    if query.lower() in ['quit', 'exit']:
        print("Ending chat. Goodbye!")
        break

    # 2. Add your new message to the history
    chat_history.append({"role": "user", "content": query})

    # 3. Send the ENTIRE history to Gemma
    stream = client.chat.completions.create(
        model="google/gemma-3-4b:2",
        messages=chat_history,
        stream=True,
        temperature=0.7
    )

    print("Gemma: ", end="")

    # 4. Stream the response back
    assistant_response = ""
    for chunk in stream:
        if chunk.choices[0].delta.content:
            content = chunk.choices[0].delta.content
            print(content, end="", flush=True)
            assistant_response += content

    print("\n")

    # 5. Add Gemma's response to the history so it remembers it for next time!
    chat_history.append({"role": "assistant", "content": assistant_response})
```
