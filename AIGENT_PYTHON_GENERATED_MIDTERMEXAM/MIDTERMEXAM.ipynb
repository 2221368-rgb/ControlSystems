```python
import control as ctrl
import numpy as np
import matplotlib.pyplot as plt


def analyze_system(num, den, title):

    # Create transfer function
    G = ctrl.TransferFunction(num, den)

    print("\n=================================")
    print(f"Analysis for {title}")
    print("=================================")

    # Display Transfer Function
    print("\nTransfer Function:")
    print(f"G(s) = {G}")

    # Poles and Zeros
    poles = ctrl.poles(G)
    zeros = ctrl.zeros(G)

    print(f"\nPoles: {poles}")
    print(f"Zeros: {zeros}")

    # Step Response
    t, y = ctrl.step_response(G)

    plt.figure(figsize=(10,5))
    plt.plot(t, y, linewidth=2)
    plt.title(f"Step Response - {title}")
    plt.xlabel("Time (s)")
    plt.ylabel("Output")
    plt.grid(True)
    plt.show()

    # Pole Zero Map
    plt.figure(figsize=(6,6))
    ctrl.pzmap(G, grid=True)
    plt.title(f"Pole-Zero Map - {title}")
    plt.show()


# ===============================
# Equation 1
# (s + 1) / (s^4 + 4s^3 + 7s^2 + 6s + 3)
# ===============================

analyze_system(
    [1,1],
    [1,4,7,6,3],
    "Equation 1"
    )


# ===============================
# Equation 2
# 1 / (s^4 + 2s^3 + 3s^2 + 2s)
# ===============================

analyze_system(
    [1],
    [1,2,3,2,0],
    "Equation 2"
    )


# ===============================
# Equation 3
# (4s^2 + 2s + 6) / (16s^4 + 24s^3 + 52s^2 + 36s)
# ===============================

analyze_system(
    [4,2,6],
    [16,24,52,36,0],
    "Equation 3"
    )
```
