A transfer function has the form:

G(s) = N(s)/D(s)

where:
N(s) = numerator polynomial  
D(s) = denominator polynomial


```python
num = [10, 20]        # Numerator coefficients  → 10s + 20
den = [1, 15, 50, 0]  # Denominator coefficients → s^3 + 15s^2 + 50s + 0
```

```python
import control as ctrl
import matplotlib.pyplot as plt

G = ctrl.tf(num, den)
```

```python
print(G)
```

```python
t, y = ctrl.step_response(G)
plt.plot(t, y)
plt.xlabel("Time (s)")
plt.ylabel("Output")
plt.title("Step Response of G(s) = 10(s+2) / [ s(s+5)(s+10) ]")
plt.grid(True)
plt.show()
```
