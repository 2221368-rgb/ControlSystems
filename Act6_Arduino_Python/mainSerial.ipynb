```python
import serial
import time

# Ensure your Arduino Serial Monitor is CLOSED before running this
try:
    arduino = serial.Serial(port='COM6', baudrate=9600, timeout=1)
    time.sleep(2) # Wait for Arduino to reset
    print("Successfully connected to COM6")

    while True:
        command = input("Enter 1 to turn ON, 0 to turn OFF, q to quit: ")
        
        if command == '1':
            arduino.write(b'1')
            print("Command Sent: LED ON")
        elif command == '0':
            arduino.write(b'0')
            print("Command Sent: LED OFF")
        elif command == 'q':
            print("Exiting...")
            break
            
    arduino.close()
    print("Port closed.")

except Exception as e:
    print(f"Failed to connect: {e}")
```
